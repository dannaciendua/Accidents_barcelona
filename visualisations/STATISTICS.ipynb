```python
# Define criteria for dangerous accidents

dangerous_conditions = [
    'Death (within 24 hours of accident)',
    'Serious injury: Hospitalization over 24 hours',
    'Death (after 24 hours of accident)'
]

# Filter for dangerous and non-dangerous accidents in both datasets
dangerous_accidents_2023 = data2023[data2023['Victimization_Description'].isin(dangerous_conditions)]
non_dangerous_accidents_2023 = data2023[~data2023['Victimization_Description'].isin(dangerous_conditions)]

dangerous_accidents_2020 = data2020[data2020['Victimization_Description'].isin(dangerous_conditions)]
non_dangerous_accidents_2020 = data2020[~data2020['Victimization_Description'].isin(dangerous_conditions)]
```

```python
contingency_table_month_df = pd.DataFrame({
    'Month': ['January', 'June'],
    'Dangerous': [20, 23],
    'Non-Dangerous': [748, 496]
})

print(contingency_table_month_df)
```

```python
odds_ratio, p_value = fisher_exact(contingency_table_month)
print(f"Odds Ratio: {odds_ratio}, P-value: {p_value}")
```

The Odds Ratio (0.59) means that accidents in the most popular month of 2023 were less common than in 2020. Specifically, for every 10 accidents in the most popular month of 2020, there were about 6 accidents in 2023.

P-value (0.11): The p-value is greater than 0.05, meaning the difference in accident counts for the most popular month between 2020 and 2023 is not statistically significant. There's insufficient evidence to conclude that the observed difference is not due to chance.


Odds are similar to chances but written as a ratio. For example, if there are 3 accidents and 7 non-accidents, the odds are 3:7.
The Odds Ratio compares two sets of odds (e.g., accidents in 2020 vs. 2023)


Отношение шансов (Odds Ratio, OR) — это показатель, который сравнивает вероятность наступления события в одной группе с вероятностью его наступления в другой группе


Шансы — это соотношение вероятности наступления события к вероятности его ненаступления. Например, если 3 аварии произошли, а 7 — нет, то шансы равны 3:7.
