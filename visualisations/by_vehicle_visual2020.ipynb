```python
# Group data by Vehicle_Brand and Vehicle_Color, then count occurrences
vehicle_counts = accidents_by_vehicle2020.groupby(['Vehicle_Brand', 'Vehicle_Color']).size().reset_index(name='Count')

# Sort by 'Count' in descending order to get the most common combinations
most_common_vehicles = vehicle_counts.sort_values(by='Count', ascending=False).head(10)

# Display the top 10 most common vehicle-brand-color combinations
print(most_common_vehicles)

# Plot the most common combinations with pastel colors
plt.figure(figsize=(10, 6))
sns.barplot(
    data=most_common_vehicles, 
    x='Count', 
    y=most_common_vehicles['Vehicle_Brand'] + " - " + most_common_vehicles['Vehicle_Color'],
    palette='pastel'  # Set pastel color palette
)

# Customize the plot
plt.title('Top 10 Most Common Vehicle Brands and Colors in Accidents 2020', fontsize=16)
plt.xlabel('Count', fontsize=12)
plt.ylabel('Vehicle Brand - Color', fontsize=12)
plt.grid(axis='x', linestyle='--', alpha=0.7)
plt.tight_layout()

# Show the plot
plt.show()
```

```python
# Filter the data for known brands and models
filtered_data = accidents_by_vehicle2020[
    (accidents_by_vehicle2020['Vehicle_Brand'] != 'Unknown') &
    (accidents_by_vehicle2020['Vehicle_Model'] != 'Unknown')
]

# Count the occurrences of each vehicle brand and model
brand_model_counts = filtered_data.groupby(['Vehicle_Brand', 'Vehicle_Model']).size().reset_index(name='Count')

# Sort by count and select the top 10
top_10_brand_models = brand_model_counts.sort_values(by='Count', ascending=False).head(10)

# Plot a horizontal bar chart with pastel colors
plt.figure(figsize=(12, 8))
plt.barh(
    top_10_brand_models['Vehicle_Brand'] + " " + top_10_brand_models['Vehicle_Model'],
    top_10_brand_models['Count'],
    color=sns.color_palette('pastel', len(top_10_brand_models)),  # Pastel color palette
    edgecolor='black'
)

# Add labels and title
plt.title('Top 10 Vehicle Brands and Models Involved in Accidents 2020', fontsize=16)
plt.xlabel('Number of Vehicles', fontsize=12)
plt.ylabel('Vehicle Brand and Model', fontsize=12)
plt.gca().invert_yaxis()  # Invert y-axis for better readability
plt.grid(axis='x', linestyle='--', alpha=0.7)
plt.tight_layout()

# Show the plot
plt.show()
```
