```python
# Group by 'District_Name', calculate the count, and sort by 'Accident_Count' in descending order
district_accidents = (
    accidents_by_cause2020.groupby('District_Name')
    .size()
    .reset_index(name='Accident_Count')
    .sort_values(by='Accident_Count', ascending=False)
)


district_accidents
```

```python
district_accidents = district_accidents[district_accidents['District_Name'] != 'Desconegut']
```

```python
least_accidents_by_district = district_accidents.sort_values(by='Accident_Count', ascending=False)
```

```python
least_accidents_by_district
```

```python
num_districts = len(least_accidents_by_district['District_Name'])
distinct_pastel_colors = plt.cm.tab20(np.linspace(0, 1, num_districts))

plt.figure(figsize=(10, 6))
plt.barh(
    least_accidents_by_district['District_Name'], 
    least_accidents_by_district['Accident_Count'], 
    color=distinct_pastel_colors
)
plt.title("Districts with Least Accidents")
plt.xlabel("Number of Accidents")
plt.ylabel("Districts")
plt.tight_layout()
plt.show()
```

```python
#here we can add some details for example accidents by area or 5 most dangerous areas
import folium
from folium.plugins import HeatMap

# Find the 5 most dangerous areas by district
most_dangerous_areas = (
    accidents_by_cause2020.groupby('District_Name')
    .size()
    .reset_index(name='Accident_Count')
    .sort_values(by='Accident_Count', ascending=False)
    .head(5)
)

# Merge with the original dataset to get coordinates for the top 5 districts
top_5_areas_data = accidents_by_cause2020.merge(
    most_dangerous_areas[['District_Name']],
    on='District_Name',
    how='inner'
)

# Create the map centered on Barcelona
map_barcelona = folium.Map(location=[41.3851, 2.1734], zoom_start=12)

# Add HeatMap for the top 5 areas
map_barcelona.add_child(
    HeatMap(top_5_areas_data[['Latitude', 'Longitud']].dropna())
)

# Add additional markers for context (optional)
for _, row in most_dangerous_areas.iterrows():
    folium.Marker(
        location=[row['Latitud'], row['Longitud']],  # Replace with appropriate lat/long if available
        popup=f"{row['District_Name']}: {row['Accident_Count']} accidents",
        icon=folium.Icon(color='red', icon='info-sign')
    ).add_to(map_barcelona)

# Display or save the map
map_barcelona.save("top_5_dangerous_areas_map.html")
map_barcelona
```
